```python
def equalisehistogram(inputimage):
    img_to_yuv = cv2.cvtColor(inputimage,cv2.COLOR_BGR2YUV)
    img_to_yuv[:,:,0] = cv2.equalizeHist(img_to_yuv[:,:,0])
    hist_equalization_result = cv2.cvtColor(img_to_yuv, cv2.COLOR_YUV2BGR)
    return hist_equalization_result

def claheapproach(img,cliplimit = 2.0,tilesize = (8,8)):
    clahe=cv2.createCLAHE(clipLimit= cliplimit, tileGridSize=tilesize)
    lab=cv2.cvtColor(img, cv2.COLOR_BGR2LAB)  # convert from BGR to LAB color space
    l,a,b=cv2.split(lab)  # split on 3 different channels
    l2=clahe.apply(l)  # apply CLAHE to the L-channel
    lab=cv2.merge((l2,a,b))  # merge channels
    outputimage=cv2.cvtColor(lab, cv2.COLOR_LAB2BGR)  # convert from LAB to BGR
    return outputimage


from PIL import ImageEnhance
from PIL import Image

def Pillowenhancer(im,factor = 1.1):
    im = Image.fromarray(im)
    enhancer = ImageEnhance.Contrast(im)
    im_output = enhancer.enhance(factor)
    im_array = np.array(im_output)
    return im_array
```

```python
m1 = claheapproach(matched)
m2 = equalisehistogram(matched)
m3 = Pillowenhancer(matched)

fig, (ax1, ax2,ax3,ax4) = plt.subplots(nrows=1, ncols=4, figsize=(10, 4),
                                    sharex=True, sharey=True)
for aa in (ax1, ax2, ax3):
    aa.set_axis_off()

ax1.imshow(matched)
ax1.set_title('og')
ax2.imshow(m1)
ax2.set_title('clahe')
ax3.imshow(m2)
ax3.set_title('equalize')
ax4.imshow(m3)
ax4.set_title('pillow')
plt.tight_layout()
plt.show()
```
